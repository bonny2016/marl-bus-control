```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

def plot_dir(csv_files_directory):
    # Load all CSV files
    csv_files = glob.glob(csv_files_directory)

    # Create a plot
    plt.figure(figsize=(10, 6))

    # Loop through each file and plot the trajectory
    for file in csv_files:
        # Load the CSV file into a DataFrame
        df = pd.read_csv(file)

        # Assume the CSV files have columns named 'time' and 'loc' for time and location
        time = df['time']/3600
        location = df['loc']
        
        plt.plot(time, location, color='blue', linewidth=2)
        # Plot the bus trajectory
#         plt.plot(time, location, label=f'Bus {file}')

    # Add labels and title
    plt.xlabel('Time')
    plt.ylabel('Location')
#     plt.title('Bus Trajectories')
    # plt.legend()  # Add a legend to distinguish buses
    plt.grid(False)

    # Show the plot
    plt.show()
    
```

```python
csv_files_directory = "./vis/visCAAC_q_v_cross_attention_crossq/*.csv"
plot_dir(csv_files_directory)
```

```python
csv_files_directory = "./vis/visnc/*.csv"
plot_dir(csv_files_directory)
```

```python
def plot_two_dirs(csv_files_directory_1, csv_files_directory_2):
    # Load all CSV files from both directories
    csv_files_1 = glob.glob(csv_files_directory_1)
    csv_files_2 = glob.glob(csv_files_directory_2)

    # Create subplots: one for each set of CSV files
    fig, axs = plt.subplots(1, 2, figsize=(15, 6))  # 1 row, 2 columns

    # Plot the first set on the left
    for file in csv_files_1:
        df = pd.read_csv(file)
        time = df['time'] / 3600.0  # Convert seconds to hours
        location = df['loc']
        axs[0].plot(time, location, color='blue', linewidth=2)

    axs[0].set_xlabel('Time (Hours)')
    axs[0].set_ylabel('Location')
    axs[0].set_title('Bus Trajectories - No Control')
    axs[0].grid(False)

    # Plot the second set on the right
    for file in csv_files_2:
        df = pd.read_csv(file)
        time = df['time'] / 3600.0  # Convert seconds to hours
        location = df['loc']
        axs[1].plot(time, location, color='blue', linewidth=2)

    axs[1].set_xlabel('Time (Hours)')
    axs[1].set_ylabel('Location')
    axs[1].set_title('Bus Trajectories - Our Control')
    axs[1].grid(False)

    # Adjust layout and show the plot
    plt.tight_layout()
    plt.show()
```

```python
 plot_two_dirs( "./vis/visnc/*.csv", "./vis/visCAAC_q_v_cross_attention_crossq/*.csv")
```
